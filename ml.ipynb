```python
y=df['Outcome']
x=df.drop('Outcome',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x ,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
lgr=LogisticRegression()
```

```python
dt=DecisionTreeClassifier()
```

```python
lgr.fit(x_test,y_test)
dt.fit(x_test,y_test)
```

```python
print(lgr.score(x_test,y_test),lgr.score(x_train,y_train))
print(dt.score(x_test,y_test),dt.score(x_train,y_train))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lgr,x_test,y_test,cv=5)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    "C":[0.01,1,10,1000],      #regularization parameter
    'penalty':['l2','l1','elasticnet'],   #peanlty types
    'solver':['liblinear']  #optimization algorithm
}
```

```python
log_reg=LogisticRegression(max_iter=5000)
```

```python
grid_search=GridSearchCV(estimator=log_reg,param_grid=param_grid,cv=5,scoring='accuracy')
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_params=grid_search.best_params_
print("best parameters:",best_params)
```

```python
lgr=LogisticRegression(C=10,max_iter=1000,penalty='l1',solver='liblinear')
```

```python
lgr.fit(x_train,y_train)
```

```python
lgr.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score

print(confusion_matrix(y_test,lgr.predict(x_test)))
print(accuracy_score(y_test,lgr.predict(x_test)))
print(precision_score(y_test,lgr.predict(x_test)))
print(recall_score(y_test,lgr.predict(x_test)))
print(f1_score(y_test,lgr.predict(x_test)))
```
